**Cleaning the data**

```python
newfull_data = full_data.drop( columns = ['Cabin', 'Name', 'PassengerId', 'Ticket'])
```

**SGD Classifier Algorithm**

```python
#Create a dataframe for all the categorival fields to apply One Hot Encoder
categorical_only = pd.DataFrame({ 'Pclass': newfull_data['Pclass'], 
                                 'Sex': newfull_data['Sex'],
                                 'SibSp': newfull_data['SibSp'],
                                 'Parch': newfull_data['Parch'],
                                 'Embarked': newfull_data['Embarked']})
categorical_only
```

```python
columns = list(categorical_only.columns)
columns
```

One Hot Encoder

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse = False)
encoder.fit(categorical_only)
column_list = list (encoder.get_feature_names_out (columns))
column_list
```

```python
categorical_np = encoder.transform(categorical_only)
```

```python
categorical_df = pd.DataFrame(categorical_np, columns = column_list)
categorical_df
```

```python
newfull_data = newfull_data.join (categorical_df)
newfull_data
```

```python
newfull_data.drop(columns = ['Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked'], inplace = True)
newfull_data
```

```python
dataX = newfull_data.drop(columns = ['Survived'])
dataY = pd.DataFrame({ 'Survived': newfull_data['Survived']})
```

```python
from sklearn.preprocessing import StandardScaler

dataXcolumns = dataX.columns.tolist()
scaler = StandardScaler ()
dataX = scaler.fit_transform (dataX)   
dataX2 = pd.DataFrame (dataX, columns = dataXcolumns)  
```

```python
from sklearn.decomposition import PCA

dataX2columns = dataX2.columns.tolist()
pca = PCA(n_components=4)
dataX3 = pca.fit_transform(dataX2)
```

```python
from sklearn.model_selection import train_test_split

trainX, testX, trainY, testY = train_test_split(dataX3, dataY, test_size = 0.20, random_state = 11)
```

```python
from sklearn.linear_model import SGDClassifier

model = SGDClassifier (shuffle = False)
model.fit (trainX, trainY.values.ravel())
```

```python
print ('R2 for training data: ',model.score(trainX, trainY))
print ('R2 for test data: ',model.score(testX, testY))
```

```python
predictions = model.predict(testX)
result = pd.DataFrame({'predictions': predictions, 'actual': testY.Survived})
result
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
print('\n\n Confusion Matrix ::\n',confusion_matrix(testY, predictions))
```

```python
#Classification report
from sklearn.metrics import classification_report
print (classification_report(testY,predictions))
```
